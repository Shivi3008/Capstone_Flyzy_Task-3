## Storing the name of mbox file in a variable

```python
mbox_file = 'All mail Including Spam and Trash.mbox'
```

### Code for Reading the email/ Mbox File 

```python
import email
from email.policy import default
```

### Creating a mboxreader class that takes file name as inout and gives the generator 
### Like object that give the next item in the list when requested

```python
class MboxReader:#Generator type of object
    def __init__(self, filename):
        self.handle = open(filename, 'rb') #rb-Read Binary
        assert self.handle.readline().startswith(b'From ')

    def __enter__(self): #dunder methods
        return self

    def __exit__(self, exc_type, exc_value, exc_traceback):
        self.handle.close()

    def __iter__(self):
        return iter(self.__next__())

    def __next__(self):
        lines = []
        while True:
            line = self.handle.readline()
            if line == b'' or line.startswith(b'From '):
                yield email.message_from_bytes(b''.join(lines), policy=default)
                if line == b'':
                    break
                lines = []
                continue
            lines.append(line)
```

### Creating a list to store the confirmed booking emails:

```python
booking_confirmation_emails = []
with MboxReader(mbox_file) as mbox:
    for message in list(mbox):
        #after analysing the data confirmed booking mail has the string "Booking Confirmed" in it
        if "Booking Confirmed" in str(message):
            booking_confirmation_emails.append(str(message))
```

### Printing all the confirmed booking emails. Also enabling scrolling view for output.

```python
for each_mail in booking_confirmation_emails:
    print(each_mail)
    print("--------------------------------------------------------------------")
```

## Final Analysis: Total number of booking emails:

```python
print(len(booking_confirmation_emails))
```
